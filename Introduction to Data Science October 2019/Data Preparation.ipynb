# Initial Data Preprocessing

#### **<font color=blue>Notice that the SurveyDate and Price columns don't seem to be of the correct data types. Let's fix that in the next subsequent cell executions</font>**

```python
#Let's check to see if we have any duplicates first
print("We have", unprocessed_data.duplicated(subset=None, keep='first').sum(),"duplicate record(s)")
```

### If we had duplicates, we could've used the below command to get rid of them

#### df = df.drop_duplicates(subset=[list_of_column_names], keep='first') <font color=blue>[generic command]</font>
#### <font color=green> df = unprocessed_data.drop_duplicates(keep='first') </font>

```python
#What are the range of values for the SurveyDate column?
unprocessed_data.SurveyDate.unique()
```

```python
#Change the SurveyDate column to a datetime object
unprocessed_data.SurveyDate = pd.to_datetime(unprocessed_data.SurveyDate, format='%Y-%m-%d')
```

```python
#What are the range of values for the Price column?
unprocessed_data.Price.unique()
#This list is too long. We might potentially miss something by just glancing over the values
```

```python
#Let's try to filter out non-numeric values in the Price column
unprocessed_data[pd.to_numeric(unprocessed_data.Price, errors='coerce').isnull()]
```

```python
#Let's impute these values with NaN's for now
unprocessed_data.Price = unprocessed_data.Price.replace('\\N', np.NaN)
```

```python
#Let's change the data type of the column now
unprocessed_data.Price = pd.to_numeric(unprocessed_data.Price)
```

```python
#Let's see if all our data types are now in order
unprocessed_data.info()
```

# Data Preprocessing Cont'd

```python
#This is one such item that had "\N" as a price value. Let's see what other prices it was being sold for
unprocessed_data[unprocessed_data.Item == 'SHAMPOO [TX] WHITE RAIN 443 mL']
```

### <font color=green>Since we've realized that they all contain 0, we can drop these rows since they will be of no use to us in further analyses</font>

```python
# Get names of indexes for which column Item has value of 'SHAMPOO [TX] WHITE RAIN 443 mL'
indexNames = unprocessed_data[unprocessed_data.Item == 'SHAMPOO [TX] WHITE RAIN 443 mL'].index

# Delete these row indexes from the dataFrame
unprocessed_data.drop(indexNames, inplace=True)
```

```python
#Let's check to see if they're still there
unprocessed_data[unprocessed_data.Item == 'SHAMPOO [TX] WHITE RAIN 443 mL']
#No longer there
```

```python
#Here's another item that had "\N" as a price value. Let's see what other prices it was being sold for
unprocessed_data[unprocessed_data.Item == 'COOKING OIL [NT] CHEF 500 mL']
```

```python
#Let's impute the 0 values with NaN's, then find the mean
unprocessed_data.Price = unprocessed_data.Price.replace(0, np.NaN)
cooking_oil_chef_500ml_mean = unprocessed_data.groupby('Item').get_group('COOKING OIL [NT] CHEF 500 mL')['Price'].mean()
cooking_oil_chef_500ml_mean
```

```python
#Now let's impute the NaN's with the mean for each
unprocessed_data.loc[(unprocessed_data.Item == 'COOKING OIL [NT] CHEF 500 mL') & (unprocessed_data.Price.isnull()), 'Price']  = cooking_oil_chef_500ml_mean
```

```python
#HLet's see if it worked
unprocessed_data[unprocessed_data.Item == 'COOKING OIL [NT] CHEF 500 mL']
```

```python
#Remember those 8 rows we dropped earlier on that contained SHAMPOO [TX] WHITE RAIN 443 mL as the item? They're no longer 
#in the dataset, hence, we have fewer rows. Let's reset our indices accordingly
unprocessed_data.info()
```

```python
#Our indices have been reset. Let's do a quick check to confirm
unprocessed_data.reset_index(drop=False, inplace=True)
unprocessed_data.info()
```
